```python
# bonus
class Disset():

    def __init__(self, subsets):
        self.subsets = []

    def clear(self):
        self.subsets = []

    def RESET(self, n):
        self.clear()
        for i in range(n):
            self.subsets.append([i])
        print("RESET DONE")

    def __check(self, j, k):
        b = False
        for i in self.subsets:
            if j in i and k in i:
                return True
        return False
    
    
    def CHECK(self, j, k):
        p = self.__check(j, k)
        if p:
            print("YES")
        else:
            print("NO")

    
    def JOIN(self, j, k):
        # проверяем
        if self.__check(j, k):
            print(f"ALREADY {j} {k}")
        else:
            # соединяем
            tmp = [] # будущий элемент множества
            for i in range(len(self.subsets)):
                if j in self.subsets[i]:
                    tmp.append(self.subsets.pop(i))
                    break
            
            for i in range(len(self.subsets)):
                if k in self.subsets[i]:
                    tmp.append(self.subsets.pop(i))
                    break
            final_output = []
            for i in range(len(tmp[0])):
                final_output.append(tmp[0][i])
            for i in range(len(tmp[1])):
                final_output.append(tmp[1][i])
            self.subsets.append(final_output)

# main
# программа заканчивается при вводе пустой строки
sets = Disset([])
p = "placeholder"
while True:
    line = input()
    if line == '':
        break
    lst = list(line.split())
    if lst[0] == "RESET":
        sets.RESET(int(lst[1]))
    if lst[0] == "CHECK":
        sets.CHECK(int(lst[1]), int(lst[2]))
    if lst[0] == "JOIN":
        sets.JOIN(int(lst[1]), int(lst[2]))        
```

```python
# 3
class Graph:
    def __init__(self, vertices):
        self.V = vertices
        self.graph = []

    def add_edge(self, u, v, w):
        self.graph.append([u, v, w])

    def find(self, parent, i):
        if parent[i] == i:
            return i
        return self.find(parent, parent[i])

    def apply_union(self, parent, rank, x, y):
        xroot = self.find(parent, x)
        yroot = self.find(parent, y)
        if rank[xroot] < rank[yroot]:
            parent[xroot] = yroot
        elif rank[xroot] > rank[yroot]:
            parent[yroot] = xroot
        else:
            parent[yroot] = xroot
            rank[xroot] += 1

    def kruskal_algo_with_an_edge(self, a, b, c):
        result = []
        i, e = 0, 0
        self.graph = sorted(self.graph, key=lambda item: item[2])
        parent = []
        rank = []
        for node in range(self.V):
            parent.append(node)
            rank.append(0)
        while e < self.V - 1:
            if [a, b, c] in result:
                u, v, w = self.graph[i]
                i = i + 1
                x = self.find(parent, u)
                y = self.find(parent, v)
                if x != y:
                    e = e + 1
                    result.append([u, v, w])
                    self.apply_union(parent, rank, x, y)
            else:
                e = e + 1
                result.append([a, b, c])
                self.apply_union(parent, rank, a, b)
        weight = 0
        for kk in result:
            weight += kk[2]
        return weight

n, m = map(int, input().split())
g = Graph(n)
edgelist = []
for i in range(m):
    a, b, c = map(int, input().split())
    g.add_edge(a - 1, b - 1, c)
    edgelist.append([a - 1, b - 1, c])

for i in range(m):
    print(g.kruskal_algo_with_an_edge(edgelist[i][0], edgelist[i][1], edgelist[i][2]))
```
