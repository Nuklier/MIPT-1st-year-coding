```python
# 4 from scratch
class Disset():

    def __init__(self):
        self.subsets = []

    def clear(self):
        self.subsets = []

    def RESET(self, n):
        self.clear()
        for i in range(n):
            self.subsets.append([i])

    def add_set(self, n): # unneccecary
        for i in self.subsets:
            if n in i:
                return
        self.subsets.append([n])

    def check(self, j, k):
        b = False
        for i in self.subsets:
            if j in i and k in i:
                return True
        return False
    
    
    def CHECK(self, j, k):
        p = self.check(j, k)
        if p:
            print("YES")
        else:
            print("NO")

    
    def JOIN(self, j, k):
        # проверяем
        if not self.check(j, k):
            # соединяем
            tmp = [] # будущий элемент множества
            for i in range(len(self.subsets)):
                if j in self.subsets[i]:
                    tmp.append(self.subsets.pop(i))
                    break
            
            for i in range(len(self.subsets)):
                if k in self.subsets[i]:
                    tmp.append(self.subsets.pop(i))
                    break
            final_output = []
            for i in range(len(tmp[0])):
                final_output.append(tmp[0][i])
            for i in range(len(tmp[1])):
                final_output.append(tmp[1][i])
            self.subsets.append(final_output)

def cut(V, op2, op3):
    V[op2].pop(V[op2].index(op3))
    V[op3].pop(V[op3].index(op2))

n, m, k = map(int, input().split())
# пустой список смежности
V = [None] * n
for i in range(len(V)):
    V[i] = []
# ввод списка смежности
for i in range(m):
    a, b = map(int, input().split())
    V[a - 1].append(b - 1)
    V[b - 1].append(a - 1)
# ввод списка команд
command = []
for i in range(k):
    name, v1, v2 = input().split()
    v1, v2 = int(v1) - 1, int(v2) - 1
    command.append([name, v1, v2])
    if name == "cut":
        cut(V, v1, v2)
# с этого момента идем вспять

darksouls = Disset()
darksouls.RESET(n)
eldenring = []
for i in range(len(V)):
    for j in V[i]:
        darksouls.JOIN(i, j)
for fake_i in range(k):
    i = k - fake_i - 1
    if command[i][0] == "ask":
        checkresult = darksouls.check(command[i][1], command[i][2])
        if checkresult == False:
            eldenring.append("NO")
        else:
            eldenring.append("YES")
    if command[i][0] == "cut":
        darksouls.JOIN(command[i][1], command[i][2])
eldenring = eldenring[::-1]
for i in eldenring:
    print(i)
```

```python
# 4 another one

class DS():
    def __init__(self, n): # n is the initial number of sets
        self.parent = list(range(n))
        self.rank = [1] * n
    
    def get_this_one(self, one):
        if self.parent[one] != one:
            self.parent[one] = self.get_this_one(self.parent[one])
        return self.parent[one]
    
    def get_these_two(self, one, two):
        return self.get_this_one(one) == self.get_this_one(two)
    
    def frien_these_two(self, one, two):
        one = self.get_this_one(one)
        two = self.get_this_one(two)
        if one != two:
            if self.rank[one] == self.rank[two]:
                self.rank[one] += 1
                self.parent[two] = one
            elif self.rank[one] < self.rank[two]:
                self.parent[one] = two
            else:
                self.parent[two] = one

n, m, k = map(int, input().split())
for i in range(m):
    loss = input() # лень нормально чистить лишний инпут
bloodborne = DS(n)
command = []
response = []
for i in range(k):
    tmp = list(input().split())
    tmp[1], tmp[2] = int(tmp[1]) - 1, int(tmp[2]) - 1
    command.append(tmp)
for fake_i in range(k):
    i = k - fake_i - 1
    if command[i][0] == "ask":
        if bloodborne.get_these_two(command[i][1], command[i][2]):
            response.append("YES")
        else:
            response.append("NO")
    else:
        bloodborne.frien_these_two(command[i][1], command[i][2])
response = response[::-1]
for i in response:
    print(i)
```

```python
# 2 finally
n, m = map(int, input().split())
V = [0] * n
for i in range(n):
    V[i] = []
used = [0] * n
rank = [0] * n # seeing the number of incoming edges
FLAG = 42
sort = []
for i in range(m):
    t1, t2 = map(int, input().split())
    V[t1 - 1].append(t2 - 1)
    rank[t2 - 1] += 1

# mandatory dfs init
def dfs(start):
    used[start] = 1 # 1. got in
    for i in V[start]:
        if used[i] == 1:
            return False
        if used[i] == 0 and not dfs(i):
            return False
    used[start] += 1 # 2. got out
    return True

# cycle check
for i in range(n):
    if used[i] == 0:
        if not dfs(i):
            FLAG = -1

queue = [] # this queue is the major part of uniqueness checking
for i in range(n):
    if rank[i] == 0:
        queue.append(i)
while len(queue) > 0:
    if len(queue) > 1 and FLAG != -1:
        FLAG = "NO"
        break
    current = queue.pop()
    sort.append(current)
    for i in V[current]:
        rank[i] -= 1
        if rank[i] == 0:
            queue.append(i)
if FLAG == "NO": # no then
    print(FLAG)
elif len(sort) == n and FLAG != -1: # the queue check has gone right
    print("YES")
else: # then the flag should be -1, showing 42 if smth goes wrong
    print(FLAG)
```
