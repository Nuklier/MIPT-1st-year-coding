```python
# 1 проверить граф на двудольность

def dfs(start, visited, g, ok):
    if visited[start] == 0:
        visited[start] = 1
    for u in g[start]:
        ok.append(visited[u]==visited[start])
        if visited[u] == 0:
            if visited[start] == 1:
                visited[u] = 2
            elif visited[start] == 2:
                visited[u] = 1
            dfs(u, visited, g, ok)
    return ok

n, m = map(int, input().split())
G = [None] * n
for i in range(n):
    G[i] = []
for i in range(m):
    inp =list(map(int, input().split()))
    G[inp[0] - 1].append(inp[1] - 1)
    G[inp[1] - 1].append(inp[0] - 1)
visited = [0] * n
ok = dfs(1, visited, G, [])
if max(ok):
    print('NO')
else:
    print('YES')
```
