```python
%%unittest_main
# 1 разложение на множители



def div(n):
    if n < 0:
        n = -n
    a = []
    x = 2
    while x <= n:
        if n % x == 0:
            a.append(x)
            n = n // x
            x = 2
        else:
            x += 1
    return a

class TestOrdinary_1(unittest.TestCase):
    def test_div(self):
        self.assertEqual(div(6), [2, 3], "should be [2, 3]")

class TestOrdinary_2(unittest.TestCase):
    def test_div(self):
        self.assertEqual(div(34), [2, 17], "should be [2, 17]")

class TestOrdinary_3(unittest.TestCase):
    def test_div(self):
        self.assertEqual(div(111), [3, 37], "should be [3, 37]")

class TestRecurring_1(unittest.TestCase):
    def test_div(self):
        self.assertEqual(div(36), [2, 2, 3, 3], "should be [2, 2, 3, 3]")

class TestRecurring_2(unittest.TestCase):
    def test_div(self):
        self.assertEqual(div(288), [2, 2, 2, 2, 2, 3, 3], "should be [2, 2, 2, 2, 2, 3, 3]")

class TestOne(unittest.TestCase):
    def test_div(self):
        self.assertEqual(div(1), [], "should be []]")

class TestNull(unittest.TestCase):
    def test_div(self):
        self.assertEqual(div(0), [], "should be []]")

class TestSimpler_1(unittest.TestCase):
    def test_div(self):
        self.assertEqual(div(19), [19], "should be [19]")

class TestSimpler_2(unittest.TestCase):
    def test_div(self):
        self.assertEqual(div(2017), [2017], "should be [2017]")


class TestNegativeSimple(unittest.TestCase):
    def test_div(self):
        self.assertEqual(div(-7), [7], "should be [7]")

class TestNegativeComplicated(unittest.TestCase):
    def test_div(self):
        self.assertEqual(div(-48), [2, 2, 2, 2, 3], "should be [2, 2, 2, 2, 3]")

class TestHuge(unittest.TestCase):
    def test_div(self):
        self.assertEqual(div(16914490471129), [2017, 2027, 2029, 2039], "should be [2017, 2027, 2029, 2039]")

class TestHumongous(unittest.TestCase):
    def test_div(self):
        self.assertEqual(div(67334151108322421569), [2017, 2017, 2017, 2017, 2017, 2017], "should be [2017, 2017, 2017, 2017, 2017, 2017]")
```

```python
%%unittest_main
# 2 мнк
import numpy as np

def lsf(x, y):
    X = np.array(x, dtype=np.float32)
    Y = np.array(y, dtype=np.float32)
    assert len(x) == len(y) and len(x) > 1, "wrong input"
    XY = []
    for i in range(len(X)):
        t = X[i] * Y[i]
        XY.append(t)
    Xsq = []
    for i in X:
        Xsq.append(i * i)
    xmean = np.mean(X)
    ymean = np.mean(Y)
    xsqmean = np.mean(Xsq)
    xymean = np.mean(XY)

    a = (xymean - xmean * ymean) / (xsqmean - xmean * xmean)
    b = (xsqmean * ymean - xmean * xymean)/(xsqmean - xmean * xmean)
    return [round(float(a), 3), round(float(b), 3)]

class Test_yISx(unittest.TestCase):
    def test_lsf(self):
        self.assertEqual(lsf([1, 2], [1, 2]), [1.0, 0.0], "should be [1.0, 0.0]")

class Test_yISxMultMinus1Plus3(unittest.TestCase):
    def test_lsf(self):
        self.assertEqual(lsf([1, 2], [2, 1]), [-1.0, 3.0], "should be [-1.0, 3.0]")

class Test_float_yIsx(unittest.TestCase):
    def test_lsf(self):
        self.assertEqual(lsf([1.5, 2.5, 3], [1.5, 2.5, 3.0]), [1.0, 0.0], "should be [1.0, 0.0]")

class Test_yISxMultMinus1Frac2Minus1(unittest.TestCase):
    def test_lsf(self):
        self.assertEqual(lsf([0, 1.5, -1], [-1, -1.75, -0.5]), [-0.5, -1.0], "should be [-0.5, -1.0]")

class Test_RealLSF_1(unittest.TestCase):
    def test_lsf(self):
        self.assertEqual(lsf([1, 1.5, 2, 6], [2, 3, 5, 7]), [0.884, 1.928], "should be [0.884, 1.928]")

class Test_RealLSF_2(unittest.TestCase):
    def test_lsf(self):
        self.assertEqual(lsf([-4, -3, 1, 2, 5], [1, 2, 2, 1, 2]), [0.044, 1.591], "should be [0.044, 1.591]")

class Test_Real_Real(unittest.TestCase):
    def test_lsf(self):
        self.assertEqual(lsf([0.092, 0.109, 0.139, 0.088, 0.106, 0.157, 0.190],
                        [9.196, 11.253, 13.552, 17.061, 26.015, 32.307, 40.535]),
                        [249.066, -9.93], "should be [249.066, -9.93]")
```
