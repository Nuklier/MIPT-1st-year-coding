```python
# 1 - сам класс, задачи дальше
class vector():

    def __init__(self, string):
        st = string.replace("{", "")
        st = st.replace("}", "")
        st = st.replace(",", "")
        lst = list(map(int, st.split(' ')))
        assert isinstance(lst[0], int) and isinstance(lst[1], int) and isinstance(lst[2], int)
        # assert для float не писал, вводить int
        self.x = lst[0]
        self.y = lst[1]
        self.z = lst[2]
    
    def show(self):
        print(f"({self.x} {self.y} {self.z})")
    
    def __abs__(self):
        return (self.x ** 2 + self.y ** 2 + self.z ** 2) ** 0.5

    def __add__(self, other):
        sm = vector(f"{self.x + other.x}, {self.y + other.y}, {self.z + other.z}")
        return sm

    def __sub__(self, other):
        sb = vector(f"{self.x - other.x}, {self.y - other.y}, {self.z - other.z}")
        return sb
    
    def scalar_mult(self, other):
        scalar = self.x * other.x + self.y * other.y + self.z * other.z
        return scalar
    
    def vec_mult_abs(self, other):
        detplus = self.y * other.z + self.x * other.y + self.z * other.x
        detminus = self.y * other.x + self.x * other.z + self.z * other.y
        return detplus - detminus
    
    def const_mult(self, other):
        return vector(f"{self.x * other}, {self.y * other}, {self.z * other}")

# 1.1 - центр масс    
"""
a = []
for i in range(5): # 5 - тк вводил для теста 5, а получение n с input не написал
    v = vector(input())
    a.append(v)
summed = vector("{0, 0, 0}")
for i in a:
    summed = summed + i
summed.x = summed.x / len(a)
summed.y = summed.y / len(a)
summed.z = summed.z / len(a)
summed.show()
# вывел (0.4 2 2) для ввода:
#{0, 0, 0}
#{0, 0, 4}
#{0, 4, 0}
#{0, 4, 4}
#{2, 2, 2}
# что правда
"""

# 1.2 - наибольшая площадь
"""
a = []
for i in range(5):
    v = vector(input())
    a.append(v)
smax = 0
for i in range(len(a) - 2):
    for j in range(i + 1, len(a) - 1):
        for k in range(j + 1, len(a)):
            v1 = a[i] - a[j]
            v2 = a[i] - a[k]
            S = v1.vec_mult_abs(v2) / 2
            if S > smax:
                smax = S
print(smax)
# для ввода 
# {0, 0, 0}
# {2, 0, 0}
# {2, 2, 0}
# {4, 0, 0}
# {4, 4, 0}
# вывел 8, что правда
"""
```
